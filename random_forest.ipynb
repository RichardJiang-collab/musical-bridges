```python
import pandas as pd
```

```python
data = pd.read_csv("music_mental_survey_results.csv")
data
```

```python
data.columns
```

```python
def print_unique_items_in_columns():
    for column in data.columns:
        unique_values = data[column].unique()
        print(f"Column: {column}")
        print(f"Unique values: {unique_values}")
        print("-" * 30)
```

```python
"""
Useless Columns for Predictions:
--------------------------------
1. Timestamp
2. Primary streaming service
3. Hours per day
4. Permissions
5. Music effects
"""
columns_to_remove = ["Timestamp", "Primary streaming service", "Hours per day", "Permissions", "Music effects"]
data = data.drop(columns=columns_to_remove)

#* Predict What - Favorite Genre
data
```

```python
empty_value_counts = data.isna().sum()

print("Columns with empty values and their counts: ")
print(empty_value_counts[empty_value_counts > 0])
```

```python
"""
1. Age - Fill with mean value
2. While working - Use Imputer to determine
3. Instrumentalist - Use Imputer to determine
4. Foreign languages - Use Imputer to determine
5. BPM - Fill with mean value
"""
from sklearn.impute import SimpleImputer
data["Age"] = data["Age"].fillna(data["Age"].mean())
data["BPM"] = data["BPM"].fillna(data["BPM"].mean())

nan_columns = ["While working", "Instrumentalist", "Foreign languages", "Composer"]
categorial_imputer = SimpleImputer(strategy="most_frequent")
for nan_column in nan_columns:
    data[[nan_column]] = categorial_imputer.fit_transform(data[[nan_column]])

data.isna().sum()
```

```python
# Print the specified columns
print(data[["While working", "Instrumentalist", "Foreign languages", "Composer"]])
```

```python
#* Encoding
# from sklearn.preprocessing import LabelEncoder
# label_encoder = LabelEncoder()
yes_no_columns = ["While working", "Instrumentalist", "Composer", "Exploratory", "Foreign languages"]
frequency_columns = ["Frequency [Classical]", "Frequency [Country]", "Frequency [EDM]", "Frequency [Folk]", "Frequency [Gospel]", "Frequency [Hip hop]", "Frequency [Jazz]", "Frequency [K pop]", "Frequency [Latin]", "Frequency [Lofi]", "Frequency [Metal]", "Frequency [Pop]", "Frequency [R&B]", "Frequency [Rap]", "Frequency [Rock]", "Frequency [Video game music]"]

for column in yes_no_columns:
    data[column] = data[column].map({'No':0, 'Yes':1})
for column in frequency_columns:
    data[column] = data[column].map({'Never':0, 'Rarely':1, 'Sometimes':2, 'Very frequently':3})

# data['Fav genre'] = label_encoder.fit_transform(data['Fav genre'])
# print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
```
